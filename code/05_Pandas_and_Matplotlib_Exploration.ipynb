```python
import pandas as pd
import numpy as np
import random

# Set seed for reproducibility
np.random.seed(42)
random.seed(42)

# Generate random data for 100 rows
cities = ['CityA', 'CityB', 'CityC', 'CityD']
days = ['Sunday','Monday', 'Tuesday','Wednesday','Thursday','Friday','Saturday']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

data = {
    'City': [random.choice(cities) for _ in range(1000)],
    'Day': [random.choice(days) for _ in range(1000)],
    'Month': [random.choice(months) for _ in range(1000)],
    'Temperature': np.random.uniform(0, 30, 1000)  # Random temperatures between 0 and 30
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV file
df.to_csv('temperature.csv', index=False)
```
