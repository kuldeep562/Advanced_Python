```python
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import RobustScaler
```

```python
wine = datasets.load_wine()
X = wine.data
y = wine.target
```

```python
columns = wine.feature_names
df = pd.DataFrame(X, columns=columns)
```

```python
df.head()
```

# Normalization using MinMaxScalar

```python
min_max_scaler = MinMaxScaler()
X_normalized = min_max_scaler.fit_transform(X)
```

```python
df_normalized = pd.DataFrame(X_normalized, columns=columns)
```

```python
df_normalized.head()
```

# Standardization using StandardScalar

```python
standard_scaler = StandardScaler()
X_standardized = standard_scaler.fit_transform(X)
```

```python
df_standardized = pd.DataFrame(X_standardized, columns=columns)
```

```python
df_standardized.head()
```

# Nomalization using normalizer

```python
scaler = Normalizer()
scaled_data = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_data, columns=columns)
```

```python
scaled_df.head()
```

# Robust Scaling similar to standardization

```python
scaler_rob = RobustScaler()
scaled_data = scaler_rob.fit_transform(X)
scaled_df = pd.DataFrame(scaled_data, columns=columns)
```
