```python
from sklearn.model_selection import train_test_split
```

```python
new_data=data[["flavanoids","proline","color_intensity","hue","alcohol","od280/od315_of_diluted_wines"]]
xtrain, xtest, ytrain, ytest = train_test_split(new_data,K, test_size=0.33, random_state=56)
```

```python
from sklearn.svm import SVC
```

```python
MK=SVC()
MK.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ypred=MK.predict(xtest)
print(accuracy_score(ytest,ypred))
```

```python
def svm_kernels(feat,target):
    xtrain, xtest, ytrain, ytest = train_test_split(new_data,K, test_size=0.33, random_state=62)
    kernels = ['linear', 'poly', 'rbf', 'sigmoid']
    accuracy=[]
    for i in kernels:
        svc=SVC(kernel=i)
        svc.fit(xtrain,ytrain)
        y_pred=svc.predict(xtest)
        acc=accuracy_score(ytest,y_pred)
        accuracy.append(acc)
        
    return accuracy     
accurac=svm_kernels(new_data,K)        
```

```python
plt.figure(figsize=(5, 5))
plt.bar(np.arange(len(accurac)), accurac)
plt.xticks(np.arange(len(accurac)), ['Linear', 'Poly', 'RBF', 'Sigmoid'])
plt.xlabel('SVM Kernels')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
dx=MinMaxScaler()
dx1=dx.fit_transform(data)
data=pd.DataFrame(dx1,columns=data.columns)
data
```

```python
y=data['Target']
data=data.drop(['Target'],axis=1)
```

```python
x=data
```

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
xd=mutual_info_regression(x,y,random_state=69)
xs=pd.DataFrame(xd,data.columns)
xs.plot(kind='barh',color='red')
plt.show
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data,y,random_state=42,test_size=0.3)
```

```python
from sklearn.svm import SVR
```

```python
dx=SVR()
dx.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import r2_score
```

```python
ypred=dx.predict(xtest)
r2_score(ytest,ypred)
```

```python
def svm_kernels(feat,target):
    xtrain, xtest, ytrain, ytest = train_test_split(feat,target, test_size=0.3, random_state=42)
    kernels = ['linear', 'poly', 'rbf', 'sigmoid']
    accuracy=[]
    for i in kernels:
        svr=SVR(kernel=i)
        svr.fit(xtrain,ytrain)
        y_pred=svr.predict(xtest)
        acc=r2_score(ytest,y_pred)
        accuracy.append(acc)
        
    return accuracy     
accuracies=svm_kernels(data,y)   
print(accuracies)    
```

```python
plt.figure(figsize=(10, 10))
plt.bar(np.arange(len(accuracies)), accuracies)
plt.xticks(np.arange(len(accuracies)), ['Linear', 'Poly', 'RBF', 'Sigmoid'])
plt.xlabel('SVM Kernels')
plt.ylabel('Accuracy')
plt.show()
```
